```python
df.plot(title='Daily Minimum Temperature',figsize=(10,8),legend=None)
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

fig,axes=plt.subplots(2,1,figsize=(10,8))
plot_acf(df['Temp'],lags=30,ax=axes[0],title='Autocorrelation (ACF)')
plot_pacf(df['Temp'],lags=30,ax=axes[1],title='Partial Autocorrelation (PACF)')
plt.tight_layout()
plt.show()
```
