```python
import datetime
def parser(x):
    return datetime.datetime.strptime('190'+x,'%Y-%m')
```

```python
series=read_csv('shampoo_sales.csv',header=0,parse_dates=[0],index_col=0,date_parser=parser)
series
```

```python
series.plot()
```

```python
series1=read_csv('shampoo_sales.csv',header=0,parse_dates=[0],index_col=0,date_parser=parser)
series1
```

```python
upsampled=series1.resample('D')
```

```python
interpolated=upsampled.interpolate(method='linear')
```

```python
interpolated.head(32)
```

```python
interpolated=upsampled.interpolate(method='quadratic',order=2)
interpolated.head(32)
```

```python
interpolated.plot()
```

```python
resample=series1.resample('Q')
quarterly_mean_sales=resample.mean()
quarterly_mean_sales.head()
```

```python
quarterly_mean_sales.plot()
```

```python
resample=series1.resample('A')
quarterly_mean_sales=resample.mean()
quarterly_mean_sales.head()
```

```python
quarterly_mean_sales.plot()
```
