```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
custom_column_names=['WSR0','WSR1']
df=pd.read_csv('eighthr.data.csv',header=None,parse_dates=[0],index_col=[0],na_values=['?'],names=custom_column_names+list(range(3,74)))
df
```

```python
df=df.dropna()
df
```

```python
df['WSR0']=df['WSR0'].astype(float)
df
```

```python
fig,axes=plt.subplots(2,1,figsize=(14,8))
df['WSR1'].plot(ax=axes[0],title='WSR0')
df['WSR0'].plot(ax=axes[1],title='WSR1')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
result=adfuller(df['WSR0'])
print(result)
if result[1]<0.05:
    print("It is stationary")
else:
    print("It is non stationary")
```

```python
result=adfuller(df['WSR1'])
print(result)
if result[1]<0.05:
    print("It is stationary")
else:
    print("It is non stationary")
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.statespace.varmax import VARMAX
```

```python
print('WSR0 causes WSR1')
print('--------------')
granger1=grangercausalitytests(df[['WSR0','WSR1']],2)
print('WSR1 causes WSR0')
print('--------------')
granger1=grangercausalitytests(df[['WSR1','WSR0']],2)
```

```python
from statsmodels.tsa.api import VAR
from sklearn.model_selection import train_test_split
train,test = train_test_split(df, test_size=0.2, shuffle=False)
model = VAR(train)
sortedmodel = model.select_order(maxlags=10)
sortedmodel.summary()
```

```python
var_model=VARMAX(train,order=(10,0),enforce_stationarity=True)
fitmodel=var_model.fit()
fitmodel.summary()
```

```python
n_forecart=12
pred=fitmodel.get_prediction()
preds=pred.prediction_mean
```

```python
preds.columns=['WSR0 Predictions','WSR1 predictons']
preds
```

```python
testvspred=pd.concat([train,preds],axis=1)
testvspred
```

```python
testvspred[['WSR0','WSR0 Predictions']].plot()
```

```python
testvspred[['WSR1','WSR1 Predictions']].plot()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(testvspred['WSR1'],testvspred['WSR1 predictions'])
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(testvspred['WSR0'],testvspred['WSR0 predictions'])
```
