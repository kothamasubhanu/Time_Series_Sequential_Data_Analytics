```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
values=np.random.randn(100)
values
```

```python
dates=pd.date_range(start="2023-01-01",end="2023-04-10",freq="D")
dates
```

```python
df=pd.DataFrame(values,index=dates,columns=['value'])
df
```

```python
df['value'].plot(kind='hist',bins=20,title='value')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.show()
```

```python
df['value'].plot(kind='line',figsize=(8,4),title='Value')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["value"])
plt.xlabel("Date")
plt.ylabel("value")
plt.title("Synthetic Time Series Dataset")
plt.show()
```

```python
df["year"]=df.index.year
df['month']=df.index.month
df['day']=df.index.day
df['weekday']=df.index.weekday
df
```

```python
df["lag_1"]=df["value"].shift(1)
df
```

```python
df["lag_1"]=df["value"].shift(2)
df
```

```python
df['rolling_mean_5']=df['value'].rolling(5).mean()
df['rolling_std_5']=df['value'].rolling(5).std()
df['rolling_min_5']=df['value'].rolling(5).min()
df['rolling_max_5']=df['value'].rolling(5).max()
df
```

```python
df['expanding_mean']=df['value'].expanding().mean()
df['expanding_std']=df['value'].expanding().std()
df['expanding_min']=df['value'].expanding().min()
df['expanding_max']=df['value'].expanding().max()
df
```
